```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

fashionMnist = tf.keras.datasets.fashion_mnist
(obrazki, etykiety), (obrazki2, etykiety2) = fashionMnist.load_data()

nazwy = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
bok = 7
plt.figure(figsize=(15,15))
for x in range(bok*bok):    
    plt.subplot(bok,bok,x+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(obrazki[x])
    plt.colorbar()
    plt.xlabel(nazwy[etykiety[x]])
plt.show()
```
