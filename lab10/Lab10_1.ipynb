\===============================================<br>
&nbsp;CAŁKOWANIE NUMERYCZNE METODĄ PROSTOKĄTÓW<br>
\===============================================

```
def calka(funkcja, a, b, i):
    dx = (b - a) / i
    wynik = 0
    for x in range(i):
        x = x * dx + a
        wynik += dx * eval(funkcja)
    return wynik

fx = '5*x*x+x'
a = 0.5
b = 3
i = 10
print(calka(fx, a, b, i))
```

\===============================================<br>
&nbsp;&nbsp;&nbsp;&nbsp;CAŁKOWANIE NUMERYCZNE METODĄ TRAPEZÓW<br>
\===============================================

```
def calka(funkcja, a, b, i):
    dx = (b - a) / i
    wynik = 0
    for x in range(i):
        x = x * dx + a
        fx1 = eval(funkcja)
        x += dx
        fx2 = eval(funkcja)
        wynik += 0.5 * dx * (fx1 + fx2)
    return wynik

fx = '5*x*x+x'
a = 0.5
b = 3
i = 10
print(calka(fx, a, b, i))
```

&nbsp;&nbsp;&nbsp;&nbsp;CAŁKOWANIE NUMERYCZNE METODĄ SIMPSONA<br>
\===============================================

```
def calka(funkcja, a, b, i):
  d = (b - a) / i
  wynik = 0
  for t in range(0, i, 2):
    x = a + d * 2 * t
    wynik += d * (funkcja(x) + 4 * funkcja(x + d) + funkcja(x + 2 * d)) / 3
  return wynik

fx = lambda x: 5*x*x+x
a = 0.5
b = 3
i = 10
print(calka(fx, a, b, i))
```
